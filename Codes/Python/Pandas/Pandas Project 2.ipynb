```python
import numpy as np
import pandas as pd
```

```python
# Read the Retail dataset with header information
infile = "C:\\Users\\SONY\\Pandas Project\\Retail.csv"
retail_df = pd.read_csv(infile, delimiter = ",")
retail_df
```

```python
# Read the Retail India dataset with header information
infile = "C:\\Users\\SONY\\Pandas Project\\Retail_India.csv"
retail_india_df = pd.read_csv(infile, delimiter = ",")
retail_india_df
```

```python
# Append the Retail India Products to main dataframe
retail_df = retail_df.append(retail_india_df).reset_index(drop = True)
retail_df
```

```python
# Get the Standard Deviation and Mean of Sales across all regions
groupby_vars = ["Region"]
retail_stat = pd.DataFrame(retail_df.groupby(groupby_vars).Sales.sum())
retail_stat['Standard Deviation of Sales'] = pd.DataFrame(retail_df.groupby(groupby_vars).Sales.std())
retail_stat["Mean of Sales"] = pd.DataFrame(retail_df.groupby(groupby_vars).Sales.mean())
retail_stat = retail_stat.rename(columns = {'Sales':'Total Sales'})
retail_stat.sort_values(by = ['Standard Deviation of Sales'], ascending = False).round(2)
```

```python
# Get CV across all regions
# CV = (STD / Mean)*100
retail_stat["CV of Sales"] = (retail_stat["Standard Deviation of Sales"] / retail_stat["Mean of Sales"]) * 100
retail_stat.sort_values(by = ['CV of Sales'], ascending = False).round(2)
```

```python
# Get the desprictive statistics for all the variables from retail dataframe
retail_desc_stat = pd.DataFrame(retail_df.describe())
retail_desc_stat = retail_desc_stat.drop("Stores", axis = 1)
retail_desc_stat = retail_desc_stat.transpose()
retail_desc_stat = retail_desc_stat.rename(columns = {'count':'N', 'mean':'Mean', 'std':'STD', 
                                                      'min':'MIN', 'max':'MAX', '25%':'Q1', '50%':'Q2', '75%':'Q3'})
retail_desc_stat.round(2)
```

```python
# Hyphothesis Testing

# One Sample T-test on Pandas dataframe

from scipy import stats

population_mean = 80000 # Null Hyphothesis

mean_sales = pd.Series(retail_desc_stat[:1].Mean.round(2))
mean_sales = mean_sales.get(key = 'Sales')

alpha = 0.5 # Type 1 Error Rate or Level of significance

sales_t_test = pd.DataFrame(stats.ttest_1samp(retail_df.Sales,population_mean), index = ["T_Value","P_Value"]).transpose()
sales_t_test["Alpha_Value"] = alpha
sales_t_test["Sample_Mean"] = mean_sales
sales_t_test["Population_Mean"] = population_mean
sales_t_test = sales_t_test.rename(index = {0:'Sales'})
sales_t_test = sales_t_test.reindex(columns = ['Population_Mean','Sample_Mean','T_Value','Alpha_Value','P_Value'])
sales_t_test["Decision"] = np.where(sales_t_test['P_Value'] < sales_t_test['Alpha_Value'],"Reject","Fail to Reject")
sales_t_test.round(2)
```
