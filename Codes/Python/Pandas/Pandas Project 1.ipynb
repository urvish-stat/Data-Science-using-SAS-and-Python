```python
def util_retail_aggr(in_dfn,analysis_vars,group_vars):
    
    "This function will agreegate the specific measures at given supplied hierarchy"
  
    print("Input Dataframe Name = ", in_dfn.name)
    print("Analysis Variables = ", analysis_vars)
    print("Grouping Variables = ", group_vars)    
    
    global retail_aggr
    
    if str(analysis_vars).upper() == "ALL":        
        retail_aggr = in_dfn.groupby(group_vars).sum()

    else:                    
        for key,value in in_dfn.iteritems():
            if key in analysis_vars:   
                var = [key]
                name = "mean_"+str(key)
                print(name)
                keep_vars = var + group_vars
                _temp_df = in_dfn[keep_vars]
                retail_aggr[name] = _temp_df.groupby(group_vars).mean()           
        
    return retail_aggr
```

```python
# Specify the Name for Dataframe
retail_df.name = "Retail Dataframe"

# Get the Total/Sum of all the numeric variables across Region and Subsidiary from retail Dataframe
util_retail_aggr(in_dfn = retail_df, analysis_vars = "All", group_vars = ["Region","Subsidiary"])
```

```python
# Get the first 20 Products in terms of their returns
by_vars = ["Returns"]
top_20_prod = retail_df.sort_values(by = by_vars, ascending = False).reset_index(drop = True)
top_20_prod = top_20_prod[:20]

# Calculate the discount for top 20 products
# > 50,000 - 0.5
# Between 15,000 to 50,000 - 0.3
# Between 10,000 to 15,000 - 0.2
top_20_prod["Discount"] = np.where(top_20_prod[by_vars] >= 50000, 0.5,
                                  (np.where(top_20_prod[by_vars] >= 15000, 0.3, 0.2)))

# top_20_prod["Discount"] = pd.cut(top_20_prod.Returns, [10000, 15000, 50000, np.inf], labels=[0.2,0.3,0.5])
# top_20_prod

# Calculate the Sales after discount
top_20_prod["Sales_after_discount"] = top_20_prod.Sales * top_20_prod.Discount
top_20_prod
```

```python
# Merge the Top 20 Product adjusted sales (discounted sales) with original Retail dataframe
retail_df2 = retail_df.merge(top_20_prod, how = "left").fillna(0)
retail_df2["Sales_after_discount"] = np.where(retail_df2.Discount == 0, retail_df2.Sales, retail_df2.Sales_after_discount)
retail_df2 = retail_df2.sort_values(by = ["Returns"], ascending = False)
retail_df2
```
