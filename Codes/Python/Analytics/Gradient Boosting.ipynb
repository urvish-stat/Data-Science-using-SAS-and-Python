```python
# Load the input data
import numpy as np
import pandas as pd
np.random.seed(10)
X = np.linspace(0, 2*np.pi, 100)[:, np.newaxis]
y = np.sin(X).ravel() + 1
```

```python
# Plot the initial data
import matplotlib.pyplot as plt
plt.figure(facecolor="white")
plt.scatter(X, y, c="k", label="training samples",s=30)
plt.ylabel("Y", fontsize=15)
plt.xlabel("X", fontsize=15)
plt.title("Gradient Boosted Regression - True Values",fontsize=20)
plt.legend()
plt.xlim([0,7])
plt.show()
```

```python
# Set the parameters for the Decision Tree and XGBOOST models
max_depth = 2          # maximum depth of the decision trees
n_estimators = 15      # how many steps to take
learning_rate = 1.0    # this controls how fast the model converges
```

```python
# Fit the Gradient Boosted Tree model (XGBOOST)

from sklearn.ensemble import GradientBoostingRegressor
MSE = [] # This is to hold the value of Mean Square Error at each iteration of XGBOOST

for i in range(1, n_estimators+1):    
    
    xgboost_dt = GradientBoostingRegressor(max_depth=max_depth,n_estimators=i,learning_rate=learning_rate)
    xgboost_dt.fit(X,y)
    y_hat = xgboost_dt.predict(X)
    loss = xgboost_dt.loss_(y,y_hat)
    MSE.append(loss)
    
# Plot MSE against each boosting iterations
mse_data = pd.DataFrame({"iterations" : range(1,n_estimators+1), "MSE" : MSE})
mse_data.plot.scatter(y = "MSE", x = "iterations")
plt.xlabel("Gradient Boosting Iteration")
plt.ylabel("Mean Square Error")
```
